```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import pandas as pd
import time
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.job_db
collection = db.indeed_listings
```

```python
#trying Splinter
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.indeed.com/jobs?q=data+engineer&l=Austin%2C+TX'
browser.visit(url)
browser.is_text_present('Indeed', wait_time=10)
```

```python
def click_next(browser):
    try:
        browser.is_element_present_by_css(".np")
        browser.find_by_css(".np").click()
    except:
        try:
            browser.find_by_css('#popover-x').click()
        except:
            pass
```

```python
for x in range(1, 10):
    html = browser.html
    soup = bs(html, 'html.parser')

    jobs = soup.find_all('div', class_="jobsearch-SerpJobCard unifiedRow row result clickcard")
    job_list = list(jobs)
    postings = {}
    
    for job in job_list:
        try:
        # Identify and return job title
            title = job.find('div', class_='title').text
        # Identify and return company name
            company = job.find('span', class_='company').text
        # Identify and return location
            location = job.find('div', class_='location accessible-contrast-color-location').text
        # Identify and return salary
            salary = job.find('span', class_='salaryText').text

        # Run only if title, price, and link are available
            if (title and company and location and salary):
            # Print results
                print('-------------')
                print(title)
                print(company)
                print(salary)
                print(location)

            # Dictionary to be inserted as a MongoDB document
                post ={
                   'title': title,
                   'company': company,
                   'salary': salary,
                    'location': location
                    }
                collection.insert_one(post)
                postings.append(post)
                
        except Exception as e:
            pass
            
    click_next(browser)
```
