```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
import requests
import pymongo
import pandas as pd
import time
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.indeed_db
collection = db.job_listings
```

```python
# URL of page to be scraped
url = 'https://www.indeed.com/jobs?q=data+engineer&l=Austin%2C+TX'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
# Examine the results, then determine element that contains sought info
# results are returned as an iterable list
results = soup.find_all('div', class_="jobsearch-SerpJobCard unifiedRow row result")

#Loop through returned results
for result in results:
    # Error handling
    try:
        # Identify and return job title
        title = result.find('div', class_='title').text
        # Identify and return company name
        company = result.find('span', class_='company').text
        # Identify and return location
        location = result.find('div', class_='location accessible-contrast-color-location').text
        # Identify and return salary
        salary = result.find('span', class_='salaryText').text

        # Run only if title, price, and link are available
        if (title and company and location):
            # Print results
            print('-------------')
            print(title)
            print(company)
            #print(salary)
            print(location)

            # Dictionary to be inserted as a MongoDB document
            post = {
               'title': title,
               'company': company,
               #'salary': salary,
                'location': location
            }
            
            
    except Exception as e:
        print(e)
```
